### Treinar o algoritmo PCA

```python
# Selecionar as colunas para PCA
X = df_countries.copy()

# Remover colunas não não necessárias 
X.drop(columns=['country', 'income_category'], axis=1, inplace=True)
```

```python
# Separa variáveis numéricas quantitativas
numeric_features = ['child_mort', 'exports', 'health', 'imports', 'income', 'inflation', 
                    'life_expec', 'total_fer', 'gdpp']
```

```python
# Definir Transformação
numeric_transformer = StandardScaler()
```

```python
# Criar Pre Processor de Transformação
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features)
    ]
)
```

```python
# Transformar os dados
X_transformed = preprocessor.fit_transform(X)
```

```python
X_transformed
```

```python
# Matriz de Covariância
np.cov(X_transformed)
```

```python
# Criar modelo PCA
modelo_pca = PCA(n_components=3)
```

```python
# Executar PCA
X_pca = modelo_pca.fit_transform(X_transformed)
```

```python
# Gerar um DataFrame com os componentes principais
df_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2', 'PC3'])
```

```python
# Visualiza componentes principais
df_pca.head(10)
```

```python
len(df_pca)
```

```python
# Incluir componentes principais no DataFrame original
df_countries['PC1'] = df_pca['PC1']
df_countries['PC2'] = df_pca['PC2']
df_countries['PC3'] = df_pca['PC3']
```

```python
# MOstra o DataFrame original com os componentes principais
df_countries.head(10)
```

### Visualizar Resultados

```python
# Autovalores
autovetores = modelo_pca.explained_variance_

# Autovetores
autovalores = modelo_pca.components_

print("Autovalores:")
print(autovalores)
print("\nAutovetores:")
print(autovetores)
```

```python
# Mostrar Chart 3D com os componentes principais
fig = px.scatter_3d(df_countries, x='PC1', y='PC2', z='PC3', color='income_category', 
                                  title='PCA - Componentes Principais', width=800, height=600, hover_data=['country'])
fig.show()
```

```python
# Reconstruir os dados com base mo PCA
X_recovered = modelo_pca.inverse_transform(X_pca)
```

```python
# Visualizar os dados reconstruidos
X_recovered.shape[1]
```

```python
# Calcular o erro de reconstrução
reconstruction_error = mean_squared_error(X_transformed, X_recovered)
print("Erro de Reconstrução:", reconstruction_error)
```

### Salvar o modelo

```python
joblib.dump(modelo_pca, './modelo_pca_country.pkl')

joblib.dump(preprocessor, './preprocessor_pca_country.pkl')
```
